```python
import json
from pathlib import Path
from transformers import pipeline

model = json.loads(Path("../../settings.json").read_text())["MODEL"]

pipe = pipeline("text-generation", model=model)
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
```

```python
from transformers import AutoTokenizer
t05 = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Coder-0.5B")
t15 = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Coder-1.5B")
t3 = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Coder-3B")
print(t05.vocab_size == t15.vocab_size == t3.vocab_size, len(t05) == len(t15) == len(t3))
```
